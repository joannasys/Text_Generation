In this notebook, I will be scrapping from the website https://bigbangtrans.wordpress.com. It is a website with transcripts of the popular American sitcom - The Big Bang Theory.

```python
from scrapy.selector import Selector
from scrapy.http import HtmlResponse
import requests

# target web page
url = "https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/"

# establishing the connection to the webpage
response = requests.get(url)

# status codes to understand how the target server responds to our request.
#Ex. 200 = OK, 400 = Bad Request, 403 = Forbidden, 404 = Not Found
print ('Status Code: ',response.status_code)

# pull HTML string out of requests and convert to a python string
html = response.text
html = Selector(text=html)

# obtain all pages
all_pages = []
for page in range(5,1071,1):
    next_page = html.xpath("//li[@class='page_item page-item-"+str(page)+"']/a/@href").extract_first()
    all_pages.append(next_page)

print ("Obtained total of {0} pages".format(len(all_pages)))
```

```python
import time
from selenium import webdriver

driver = webdriver.Chrome(executable_path="./chromedriver/chromedriver")

titles = []
texts = []

def extract(html):
    
    for sel in html.xpath("//div[@id='content']"):
        
        # extracting title
        title = sel.xpath("//h2[@class='title']/text()").extract()
        titles.append(title)
        
        # there are some series/episode whereby the content is in different tags
        
        series_6 = ['Series 06 Episode 01']
        
        series_7 = ['Series 07 Episode '+str(i) for i in range(1,16)]
        check_1 = ['Series 02 Episode 01']
        check_1.extend(series_7)
        
        check_2 = ['Series 02', 'Series 03', 'Series 04', 'Series 05','Series 06', 'Series 08','Series 09']
        series_7 = ['Series 07 Episode '+str(i) for i in range(16,25)]
        check_2.extend(series_7)
                
        for t in title:
            
            if any(epi in t for epi in series_6):
                text = sel.xpath("//div[@class='entrytext']/p/span/text()").extract()
                break
            
            elif any(epi in t for epi in check_1):
                text = sel.xpath("//span/text()").extract()
                break

            else:
                if any(epi in t for epi in check_2):
                    text = sel.xpath("//div[@class='entrytext']/p/text()").extract()
                    break
                else:
                    text = sel.xpath("//span/text()").extract()
                    break

            #texts.append(sel.xpath("//span/text()").extract())
            #texts.append(sel.xpath("//div[@class='entrytext']/p/text()").extract())

        texts.append(text)

for page in all_pages:
    if page is not None:
        try:
            driver.get(page)
            #response = requests.get(page)
            print ('Establishing Connection...')
            # to allow the webpage to finish loading
            time.sleep(2)
            #html = response.text
            html = driver.page_source
            print ('Obtain html...')
            html = Selector(text=html)
            extract(html)
            print ("Extracted episode",str(page))
            all_pages.pop(page)
            

        except:
            pass
        
# store data into a dictionary         
bbt = {}
bbt['titles'] = titles
bbt['texts'] = texts


import pickle

# pickle the data
f = open("bbt_2.pickle","wb")
pickle.dump(bbt,f)
f.close()

# convert into csv
import pandas as pd
pd.DataFrame(bbt).to_csv('bbt_2.csv')
```
